```python
import pandas as pd
```

```python
IN_PATH = "/shared/3/projects/newsDiffusion/data/processed/fullDataWithClustNums.tsv"
df = pd.read_csv(IN_PATH, sep="\t")
```

```python
df = df.dropna(subset=["clustNum"]) 
```

```python
grouped = df[["clustNum", "source", "content"]].groupby("clustNum").agg(list)
```

```python
grouped["clustLen"] = grouped["source"].apply(len)
```

```python
bigGroup = grouped[grouped["clustLen"] > 100000] 
```

```python
bigGroup = bigGroup.explode(["source", "content"])
```
